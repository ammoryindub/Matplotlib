```python
# Imprt libraries

import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
# first we will create couple of datasets we can use in our plotting exercise 

x = np.arange(0,99)
y = x**2
z = x**3
```

#### creating a figure object
#### adding axes to the figure canvas
#### plot axes
#### set labels and titles to figure

```python
fig = plt.figure()

axes = fig.add_axes ([0.5,0.5,2,2])
axes.plot (x,y)
axes.set_xlabel('X lable')
axes.set_ylabel('Y lable')
axes.set_title('Title')

plt.show()
```

#### creating a figure object
#### put 2 axes on it
#### ax1 located [1,1,1.5,1,1.5]
#### ax2 located [0.4,0.8,.6,.6]

```python
fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1.5, 1.5])
ax2 = fig.add_axes([0.4, 0.8, 0.6, 0.6])

ax1.plot(x,y,color='red')
ax2.plot(x,z,color ='green')

plt.show()
```

#### creating a figure object
#### put 2 axes on it, now lets try using ax2 as a zoom of part of ax2 values 
#### ax1 located [0,0,10,10]
#### ax2 located [0.4,0.8,1,1]

```python
fig = plt.figure()



ax1 = fig.add_axes([0,0,6,6])
ax2 = fig.add_axes([0.4,4.5,1,1])

# Large plot
ax1.plot(x,y,color='red')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')

# small plot
ax2.plot(x,z,color ='green')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('Zoom')

# now we set ax2 limits so we can enter the values we wanna zoom at

ax2.set_xlim(0,20)
ax2.set_ylim(200,400)

plt.show()
```

#### creating a figure object
#### using subplot to present 2 plots 
#### plot 1 shows (x,y)
#### plot 2 shows (x,z)

```python
# creating subplots

fig,axes = plt.subplots(nrows=1, ncols=2)
```

```python
# adding style to subplots

axes[0].plot(x,y,color="blue", lw=2, ls='--')
axes[1].plot(x,z,color="red", lw=1, ls='-')

fig
```

```python
# adding labels to subplots

axes[0].plot(x,y,color="blue", lw=2, ls='--')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].set_title('X squared')

axes[1].plot(x,z,color="red", lw=1, ls='-')
axes[1].set_xlabel('X')
axes[1].set_ylabel('Z')
axes[1].set_title('X cubed')

fig.tight_layout()
fig
```

#### we will establish new random arrays, and we will be trying to sue them in different plot types 
#### that will help mastering plots more

#### N: note that each type of data has a different plot style suited for presentation and visualization purposes 

```python
# we will intiate our random arrays, we will use different ways! 

x = np.random.rand(25)
y = np.linspace (0,8,25)
```

```python
x
```

```python
y
```

```python
# size and color:

sizes = np.random.uniform(0, 80, len(x))
colors = np.random.uniform(0, 80, len(x))
```

```python
# scatter plot

fig, ax = plt.subplots()

ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)

ax.set(xlim=(0, 6), xticks=np.arange(1, 6),
       ylim=(0, 6), yticks=np.arange(1, 6))

plt.show()
```

```python
# stem plot

fig, ax = plt.subplots()

ax.stem(x, y)

ax.set(xlim=(0, 6), xticks=np.arange(1, 6),
       ylim=(0, 6), yticks=np.arange(1, 6))

plt.show()
```

```python
# hist plot

fig, ax = plt.subplots()

ax.hist(x, bins=8, linewidth=0.5, edgecolor="white")

ax.set(xlim=(0, 6), xticks=np.arange(1, 6),
       ylim=(0, 6), yticks=np.linspace(0, 6, 5))

plt.show()
```

```python
# boxplot

fig, ax = plt.subplots()

VP = ax.boxplot(x, positions=[2], widths=1.5, patch_artist=True,
                showmeans=False, showfliers=False,
                medianprops={"color": "white", "linewidth": 0.5},
                boxprops={"facecolor": "C0", "edgecolor": "white",
                          "linewidth": 0.5},
                whiskerprops={"color": "C0", "linewidth": 1.5},
                capprops={"color": "C0", "linewidth": 1.5})

ax.set_xlim(0, 4)
ax.set_ylim(0, 6)
ax.set_xticks([2])
ax.set_xticklabels(['x'])
ax.plot([2], [y.mean()], marker='o', color='red')

plt.show()
```

```python
# grid plot

plt.plot(x, y, 'o-')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of x and y')
plt.grid(True)
plt.show()
```

```python
# pie plot


fig, ax = plt.subplots()

colors = plt.get_cmap('Blues')(np.linspace(0.2, 0.7, len(x)))

ax.pie(x, colors=colors, radius=3, center=(4, 4),
       wedgeprops={"linewidth": 1, "edgecolor": "white"}, frame=True)

ax.set(xlim=(0, 10), xticks=np.arange(1, 10),
       ylim=(0, 10), yticks=np.arange(1, 10))

plt.show()
```

```python
z = (y/x)**2
```

```python
# tricontour plot

fig, ax = plt.subplots()

levels = np.linspace(z.min(), z.max(), 20)

ax.plot(x, y, '+', markersize=8, color='lightgrey')
ax.tricontour(x, y, z, levels=levels)

ax.set_xlim(-3, 3)
ax.set_ylim(-3, 3)

plt.show()
```

```python
# 3D scatter plot


from mpl_toolkits.mplot3d import Axes3D


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, z, c='lightgrey', s=30, edgecolors='none')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

```python
# Triangular 3D surfaces plot


# Triangular surface coordinates
triangles = np.tri(len(x), len(y), dtype=int)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
ax.scatter(x, y, z, c='lightgrey', s=30, edgecolors='none')

# Triangular surfaces
ax.plot_trisurf(x, y, z, triangles=triangles, cmap='viridis')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

```python
# 3D wireframe plot


# Reshape the data into 2D grids
X, Y = np.meshgrid(x, y)
Z = (Y / X) ** 2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
ax.scatter(x, y, z, c='lightgrey', s=30, edgecolors='none')

# Wireframe plot
ax.plot_wireframe(X, Y, Z, color='blue')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
